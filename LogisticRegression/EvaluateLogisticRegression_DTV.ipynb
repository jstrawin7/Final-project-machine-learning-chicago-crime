### Remove Arrest column from evaluation data since we will be prediction Arrests

```python
# get data from truncated chicago crime 500 rows data file
df_predict = pd.read_csv("./Crime_last500.csv")
print(df_predict.shape)
```

```python

df_predict_arrest = pd.DataFrame([ df_predict.IUCR, df_predict.sin_time ,df_predict.cos_time, df_predict.XAxis, df_predict.YAxis]).transpose()
```

```python
df_predict_arrest.head(5)
```

## Load the saved model

```python
from sklearn.externals import joblib
lr_cv_model = joblib.load("./chicago-crime-trained-LR500model.pkl")
```

```python
X_predict = df_predict_arrest
```

```python
lr_cv_model.predict(X_predict)
```

```python
df_predict_arrest['Arrest_Predicted']=lr_cv_model.predict(X_predict)
```

```python
df_predict_arrest['Actual_Arrest'] = pd.DataFrame([df_predict.Arrest]).transpose()
```
